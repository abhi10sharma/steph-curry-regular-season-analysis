# Curry's Age Wise Stats Comparison

```python
df['Age'] = df['Age'].apply(lambda x: x.split('-')[0])
df
```

```python
age_df = df.groupby(['Age']).sum()
age_df
```

```python
plt.figure(figsize=(6,4), dpi = 150)

age = [Age for Age, df in df.groupby(['Age'])]

plt.bar(age, age_df['3P'], color = '#e1fc47', edgecolor = 'Black')

plt.title('3P Comparison by Age', fontsize = 20)

plt.xticks(age)


plt.xlabel('Age', fontsize = 15)
plt.ylabel('Total 3-Pointers', fontsize = 15)

plt.savefig('3P_comparison.png')
plt.show()
```

```python
plt.figure(figsize=(6,4), dpi = 150)

age = [Age for Age, df in df.groupby(['Age'])]

plt.bar(age, age_df['3PA'], color = '#e1fc47', edgecolor = 'Black')

plt.title('3PA Comparison by Age', fontsize = 20)

plt.xticks(age)


plt.xlabel('Age', fontsize = 15)
plt.ylabel('Total 3-Pointers', fontsize = 15)

plt.savefig('3PA_comparison.png')
plt.show()
```

```python
df_3p = age_df.filter(['3P', '3PA'])
df_3p
```

```python
from decimal import Decimal

plt.rcParams.update({'font.size': 22}) # must set in top

ax = df_3p.plot.bar(rot=0, figsize=(18,9), edgecolor = 'Black')
ax.tick_params(axis='both', which='both', labelsize=22)
ax.set_xlabel('Age',fontdict={'fontsize':30})
ax.set_ylabel('3P Attemp/Made',fontdict={'fontsize':30})
ax.legend(loc=2,fontsize=20)

plt.title("Total 3P Made and Attempts",fontdict={'fontsize':30})
plt.show()
```

# Curry's Stats Comparison Against All Teams

```python
opp_df = df.groupby(['Opp']).sum().sort_values('PTS', ascending = False)
opp_df
```

```python
plt.figure(figsize=(7,4), dpi = 300)

team = [Opp for Opp, df in df.groupby(['Opp'])]

plt.bar(team, opp_df['PTS'], color = '#44e3a9', edgecolor = 'Black')

plt.title('Team Against Scored Most Point', fontsize = 20)

plt.xticks(team, rotation = 'vertical')


plt.xlabel('Teams', fontsize = 15)
plt.ylabel('Total Points Scored', fontsize = 15)

plt.savefig('team_wise_comparison.png')
plt.show()
```
