# Curry's Age Wise Stats Comparison

```python
df['Age'] = df['Age'].apply(lambda x: x.split('-')[0])
df
```

```python
age_df = df.groupby(['Age']).sum()
age_df
```

```python
plt.figure(figsize=(6,4), dpi = 150)

age = [Age for Age, df in df.groupby(['Age'])]

plt.bar(age, age_df['3P'], color = '#e1fc47', edgecolor = 'Black')

plt.title('3-P Comparison by Age', fontsize = 20)

plt.xticks(age)


plt.xlabel('Age', fontsize = 15)
plt.ylabel('Total 3-Pointers', fontsize = 15)

plt.savefig('3P_comparison.png')
plt.show()
```

# Curry's Stats Comparison Against All Teams

```python
opp_df = df.groupby(['Opp']).sum().sort_values('PTS', ascending = False)
opp_df
```

```python
plt.figure(figsize=(7,4), dpi = 300)

team = [Opp for Opp, df in df.groupby(['Opp'])]

plt.bar(team, opp_df['PTS'], color = '#44e3a9', edgecolor = 'Black')

plt.title('Team Against Scored Most Point', fontsize = 20)

plt.xticks(team, rotation = 'vertical')


plt.xlabel('Team', fontsize = 15)
plt.ylabel('Total Points Made', fontsize = 15)

plt.savefig('team_wise_comparison.png')
plt.show()
```
